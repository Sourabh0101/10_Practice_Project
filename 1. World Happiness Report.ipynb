# Removing outlires

```python
outliers_column=["Standard Error", 'Family' , "Trust (Government Corruption)" , "Generosity" ,"Dystopia Residual"]
```

```python
for i in outliers_column:
    Q3 = df[i].quantile(0.75)
    Q1 = df[i].quantile(0.25)
    IQR = Q3-Q1
    upper_limit = Q3 + 1.5 * IQR 
    lower_limit = Q1 - 1.5 * IQR
    (df[(df[i]>upper_limit) | (df[i]<lower_limit)])
    df[i] = np.where(df[i] > upper_limit , upper_limit ,np.where(df[i] < lower_limit , lower_limit , df[i]))
```

# Selecting Random state

```python
def maxr2_score(pipe,X,y):
    max_r_score=0
    for i in range(42,100):
        X_train,X_test,y_train,y_test=train_test_split(X,y,random_state= i ,test_size=0.20)
        pipe.fit(X_train,y_train)
        y_pred=pipe.predict(X_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to",i,"is",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_i=i
    print("max r2 score corresponding to",final_i,"is",max_r_score)
    return final_i
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

step1 = ColumnTransformer(transformers =[        
           ('ohe' , OneHotEncoder(sparse=False,handle_unknown='ignore'),[0])
],remainder='passthrough')

step2 = ColumnTransformer(transformers=[
        ('std' ,StandardScaler(),slice(0,-1))
],remainder='passthrough')

step3 = LinearRegression()


pipe = Pipeline([
    ('step1',step1),
    ('step2',step2),
    ('step3',step3)
])

i=maxr2_score(pipe,X,y)
```

```python
print("Mean r2 score for Linear Regression:",cross_val_score(pipe,X,y,cv=5,scoring="r2").mean())
print("standard deviation in r2 score for Linear Regression",cross_val_score(pipe,X,y,cv=5,scoring="r2").std())
```
